```python
def test(ArrivalsWorldWeather):
  URL = ('http://apis.data.go.kr/B551177/StatusOfPassengerWorldWeatherInfo/'
       'getPassengerArrivalsWorldWeather?'
       'ArrivalsWorldWeather='+ArrivalsWorldWeathe+
       '&serviceKey=ttkbFJE3BclD7LePyeR65I5jzC9qSvjPPRhvqL%2FupX81gOanw8XAI7hEeTpTGyig13Rk8pdZo%2FT6bWWFkE%2BxCg%3D%3D'
       '&numOfRows=10'
       '&pageNo=1'
       '&from_time=0000'
       '&to_time=2400'
       '&lang=K'
       '&type=json')
  
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']
  
  airport = [item['airport'] for item in items]
  himidity = [item['himidity'] for item in items]
  senstemp = [item['senstemp'] for item in items]
  temp = [item['temp'] for item in items]
  wind = [item['wind'] for item in items]
  
  airport = pd.Series(airport)
  himidity = pd.Series(himidity)   
  senstemp = pd.Series(senstemp)
  temp = pd.Series(temp)
  wind = pd.Series(wind)

  df = pd.DataFrame({ 'airport': airport, 'himidity': himidity, 'senstemp': senstemp, 'temp': temp, 'wind': wind})
  return df
```
